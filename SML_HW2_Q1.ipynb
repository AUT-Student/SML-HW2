```python
import pandas as pd
import numpy as np
from scipy.stats import norm
from scipy.signal import fftconvolve
import matplotlib.pyplot as plt
```

```python
q1 = norm(3, 0.5)
q2 = norm(5, 1.5)
q3 = norm(7, 0.5)
```

```python
p1 = norm(loc=3, scale=0.5)
p2 = norm(loc=7, scale=0.5)
```

```python
def kullback_leibler(p_samples, q_samples):
  return sum(p_samples * np.log(p_samples/q_samples))
```

```python
x = np.linspace(0, 10, 10000)
```

```python
p_samples = 0.5*norm(loc=3, scale=0.5).pdf(x) + 0.5*norm(loc=7, scale=0.5).pdf(x)
q1_samples = q1.pdf(x)
q2_samples = q2.pdf(x)
q3_samples = q3.pdf(x)
```

```python
print("KL(p, q1) = ", kullback_leibler(p_samples, q1_samples))
print("KL(p, q2) = ", kullback_leibler(p_samples, q2_samples))
print("KL(p, q3) = ", kullback_leibler(p_samples, q3_samples))
print("KL(q1, p) = ", kullback_leibler(q1_samples, p_samples))
print("KL(q2, p) = ", kullback_leibler(q2_samples, p_samples))
print("KL(q3, p) = ", kullback_leibler(q3_samples, p_samples))
```

```python
plt.plot(x, p_samples, label="p", color="blue")
plt.plot(x, q2_samples, label=r"$q_1^*$", color="red")

plt.xlabel("x")
plt.ylabel("f")

plt.title(r"Distribution p & Distribution $q_1^*$")

plt.legend()

plt.show()
```

```python
plt.plot(x, p_samples, label="p", color="blue")
plt.plot(x, q1_samples, label="$q_2^*$", color="red")

plt.xlabel("x")
plt.ylabel("f")

plt.title(r"Distribution p & Distribution $q_2^*$")

plt.legend()

plt.show()
```
